```python
# Statistical Test for Bivaritate Analysis
# T-test process: Step-by-step -> Flight_delays

import pandas as pd
flight_delay = pd.read_csv("flight_delays_train.csv")
flight_delay
```

```python
flight_delay['DepTime'] = pd.to_numeric(flight_delay['DepTime'], errors = 'coerce')
flight_delay['Distance'] = pd.to_numeric(flight_delay['Distance'], errors = 'coerce')

# T-test [Grouping dep_delayed_15min into yes or no]
delayed_flights = flight_delay[flight_delay['dep_delayed_15min'] == 'Y']
non_delayed_flights = flight_delay[flight_delay['dep_delayed_15min'] == 'N']
```

```python
delayed_flights
```

```python
non_delayed_flights
```

```python
#T_test for Departure Time
from scipy.stats import ttest_ind 
t_stats, p_value = ttest_ind(delayed_flights['DepTime'], non_delayed_flights['DepTime'])

t_stats, p_value
```

```python
# T_test for Distance
from scipy.stats import ttest_ind
t_stats, p_value = ttest_ind(delayed_flights['Distance'], non_delayed_flights['Distance'])

t_stats, p_value
```
